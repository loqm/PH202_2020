##  1.Introduction

Rayleigh’s criterion for resolving two incoherent point sources, requiring them to be separated at least by a diffraction-limited spot size on the image plane has been the most influential measure of optical imaging resolution for over a century. The analysis of optical resolution has been reformed in terms of Fisher Information(FI) which helps in quantifying the precision of measurements and is inversely proportional to parameter estimation error. Generally, the FI of the estimation of separation $\delta$ between two spatially incoherent point sources depends on the type of measurement performed on the image plane field. In the case of direct detection of image plane intensity, the FI goes to zero as $\delta\rightarrow 0$, an effect termed as Rayleigh’s curse. Our work investigates the role of partial coherence on the sub-diffraction limit localization of two sources based on parity sorting. Partially coherent sources are sources where the coherence time or coherence length are limited by bandwidth, which might be due to some thermal noise, or by other effect. So, we study how this thing affects the region where we can’t resolve these sources for which the paper devises a setup using an appropriate spatial mode demultiplexer(SPADE) to see whether super resolution can be achieved.\
There has a debate whether SPADE can overcome Rayleigh's curse if both the sources are spatially partially coherant. Larson and Saleh(LS) model predicted that Rayleigh's curse cannot be broken for partially coherant sources. However the Tsang-Nair(TN) model argued that SPADE can break Rayleigh's curse. More importantly, TN model showed how an experiment can be done to show super-resolution for partially coherant light. Our work inherits different elements of these works and combines them
to show the breaking of Rayleigh’s curse for partially coherent light sources.The given work is based on two contrasting papers only with respect to $\delta$(for the case when $\delta\rightarrow 0$) and assume a known degree of spatial coherance $\gamma$:
1. The first paper argues that when $\gamma=-1$, TN model predicts FI being maximized, where as LS predicts FI $\rightarrow 0$.
2. When $\gamma<0$, the FI for TN model model remains bounded below max FI as found when $\gamma=-1$. But for the LS model, the FI can grow larger than the bound predicted for TN model. 

Both models agree in their predictions of FI exactly when $\gamma=0$, and qualitatively when $\gamma>0$. This paper demonstrates that Rayleigh’s curse can be broken by SPADE when $\gamma<0$, while the curse resurges for $|\gamma|=1$.

## 2.SPADE using parity sorting

### 2.1 Model and Spatial Coherance Generation 

SPADE can be performed on two spatially partially coherant sources using a parity sorter, where we decompose the mutual coherance in the symmetric(in phase) and antisymmetric(out of phase) combinations of two sources. We assume that each source is emitting $N_0$ photons on average and the total number of photons is $N_t = 2N_0$. The mutual coherance $\Gamma(x_1,x_2)$ of the LS model in the image plane is
\begin{equation*}
    \Gamma_{LS}(x_1,x_2)=\frac{N_t}{A}\sum_{k=1}^2p_k\phi^*_k(x_1)\phi_k(x_2)
\end{equation*}
Furthermore, we assume
\begin{equation*}
    p_1+p_2=1\\
    p_1-p_2=\gamma \text{ ,$\gamma$ is the effective degree of spatial coherance between the two sources.}\\
    \phi_k(x)=f_{+}(x)-e^{ik\pi}f_{-}(x) \text{ are the symmetric(k=1) and the antisymmetric(k=2) coherant modes.} \\
    f_{\pm}(x)=f\left(x\pm\frac{\delta}{2}\right) \text{ are the two point spread functions separated by $\delta$ - the
    parameter to be estimated.}
\end{equation*}
The terms even and odd modes are used interchangeably with symmetric and anti-symmetric modes respectively. The coherant modes can be generated by a mode converter using linear optical transformations, as shown in the figure of the experimental setup. We use Gaussian PSFs of width $\sigma$ such that $f(x)=\frac{exp(\frac{-x^2}{4\sigma^2})}{(2\pi\sigma^2)^{1/4}}$. The average number of photons at point at point x is given by the intensity $\Gamma(x,x)$ as
\begin{equation*}
    \Gamma(x,x)=\frac{N_t}{A}\left(p_1|\phi_1(x)|^2+p_2|\phi_2(x)|^2\right) \\
    \Gamma(x,x)=\frac{N_t}{A}\left(|f_{+}(x)|^2+|f_{-}(x)|^2+2\gamma f_{+}(x)f_{-}(x)\right)
\end{equation*}
where we have used the relations between $p_1$ and $p_2$ stated above. In contrast to the LS model, the mutual coherance function to the TN model is given by
\begin{equation*}
    \Gamma_{TN}(x_1,x_2)=\frac{N_t}{2}\sum_{k=1}^2p_k\phi^*_k(x_1)\phi_k(x_2)
\end{equation*}
where we have used the same notation for $p_k, \phi_k$ and $N_t$ is the total photon number in the interferometer. We immediately see that $\Gamma_{TN} (x_1, x_2) = \frac{A}{2}\Gamma_{LS}(x_1, x_2)$ and the two states differ by the scaled normalization factor A.

### 2.2 Experimental details

After generating partial coherance, the next step is to perform SPADE on the field described by previous equation on $\Gamma(x_1,x_2)$. We want to implement SPADE using an image inversion interferometer which can sort the input field based on its parity as illustrated in the figure below.

A $795 nm$ Gaussian beam with $\sigma=327\pm 4 nm$ is converted into either a symmetric($\phi_1$) or antisymmetric($\phi_2$) mode using linear optics, which includes a spatial light modulator(SLM). At any given time, one of the coherant modes is presented to a Michelson type image inversion interferometer, which can separate the even and odd components of the input field. One arm has a 4f system which unitly magnifies the field after double reflections and the other arm(the top arm) has a 2f system and an extra quadratic phase(not shown in the figure above) to cancel the defocus due to diffraction. Thus, this arm also helps in implementing the transformation $(x,y)\rightarrow (-x,-y)$. The combined beams from both the arms are then imaged onto a bucket detector where the power in the even and odd modes can be measured by setting the global phase difference $\theta$ between the two arms to $0$ and $\pi$ respectively. Since all modes used in this experiment are symmetric about the y axis such that $E(x,-y)=E(x,y)$, hence the interferometer works as a parity sorter in the x-direction.\
Once we send one of the coherant modes $\phi_i$ at a given time, we see that from the intensity profile of LS and TN models that each coherant mode adds on an intesity basis. For $\theta=0(\pi)$ all of the symmetric(antisymmetric) mode power will be directed towards the bucket detector, while the antisymmetric(symmetric) mode will destructively interfere at the detector. Hence for $\theta=0(\pi)$, the output is called as the even(odd) port.

### 2.3 Equations used in both LS and TN models

We now explain how to find the parity sorter’s modal weights. The intensity profile in the LS and TN models are given as
\begin{equation*}
    I_{LS}(x)=\Gamma_{LS}(x,x)=\frac{N_t}{A}\left(p_1|\phi_1(x)|^2+p_2|\phi_2(x)|^2\right) \\
    I_{TN}(x)=\Gamma_{TN}(x,x)=\frac{N_t}{2}\left(p_1|\phi_1(x)|^2+p_2|\phi_2(x)|^2\right)
\end{equation*}
Thus, the normalization constant is given as
\begin{equation*}
    A=\int_{-\infty}^{\infty}\sum_{k=1}^2p_k|\phi_k(x)|^2 dx=2(p_1+p_2+\gamma d)=2(1+\gamma d)
\end{equation*} 
such that
\begin{equation*}
    \int \Gamma(x,x) dx=N_t \\
    d=\int f_{+}(x)f_{-}(x) dx=e^{{-\delta^2}/{8\sigma ^2}}
    \end{equation*}
The bucket detector can be used to measure the photon number $n_k$in each port, which is given by
\begin{equation*}
    n_{k,LS}=\frac{N_t p_k}{A}\int|\phi_k(x)|^2 dx=\frac{2N_t p_k}{A}(1-e^{-ik\pi}d) \\
    n_{k,TN}=\frac{N_t p_k}{2}\int|\phi_k(x)|^2 dx=N_t p_k(1-e^{-ik\pi}d)
\end{equation*}
We can check the following relations of the total photon number of both the LS and TN states
\begin{equation*}
    n_{1,LS}+n_{2,LS}=N_t \\  n_{1,TN}+n_{2,TN}=N_t(1+\gamma d)
\end{equation*}
To perform an estimation on $\delta$, we calculate $P_{1,2}$: they represent the probabilities of detecting a click in the
even or odd ports for a particular $\gamma$,$\delta$ and $\sigma$, given a single photon in the image plane, irrespective of the number of photons in the object plane. For the LS model, we have
\begin{equation}
    P_{1,LS}=\frac{n_{1,LS}}{n_{1,LS}+n_{2,LS}}=\frac{2N_tp_1}{N_t A}(1+d)=\frac{1+\gamma}{2(1+\gamma d)}(1+d)\\
    P_{2,LS}=\frac{n_{2,LS}}{n_{1,LS}+n_{2,LS}}=\frac{2N_tp_2}{N_t A}(1-d)=\frac{1-\gamma}{2(1+\gamma d)}(1-d)
\end{equation}
Here we have used $p_{1,2}=\frac{1\pm\gamma}{2}$.Similarly, the TN model also corresponds to these relations and therefore $P_{1(2),LS}=P_{1(2),TN}$. They can be used to calculate FI for a parity sorter using the expression
\begin{equation*}
    F(\delta,\gamma)=\sum_{i=1,2}\frac{1}{P_i(\delta,\gamma)}\left(\frac{\partial P_i(\delta,\gamma)}{\partial \delta}\right)^2\\
    F(\delta,\gamma)=\frac{\delta^2 d^2(1-\gamma ^2)}{16 \sigma ^4 (1-d^2)(1+\gamma d)^2}
\end{equation*}

```python
import numpy as np
x=np.linspace(0.0001,5,100)
d=np.exp(-x**2/8)
Gamma=-1
y=(x**2)*(d**2)*(1-Gamma**2)/(4*(1-d**2)*(1+Gamma*d)**2)  #equation for the LS model
line1,=plt.plot(x,y,label='$\gamma=-1$')
Gamma=-0.75
y=(x**2)*(d**2)*(1-Gamma**2)/(4*(1-d**2)*(1+Gamma*d)**2)
line2,=plt.plot(x,y,label='$\gamma=-0.75$')
Gamma=-0.5
y=(x**2)*(d**2)*(1-Gamma**2)/(4*(1-d**2)*(1+Gamma*d)**2)
line3,=plt.plot(x,y,label='$\gamma=-0.5$')
Gamma=0
y=(x**2)*(d**2)*(1-Gamma**2)/(4*(1-d**2)*(1+Gamma*d)**2)
line4,=plt.plot(x,y,label='$\gamma=0$')
Gamma=0.5
y=(x**2)*(d**2)*(1-Gamma**2)/(4*(1-d**2)*(1+Gamma*d)**2)
line5,=plt.plot(x,y,label='$\gamma=0.5$')
Gamma=1
y=(x**2)*(d**2)*(1-Gamma**2)/(4*(1-d**2)*(1+Gamma*d)**2)
line6,=plt.plot(x,y,label='$\gamma=1$')
plt.legend(handles=[line1, line2, line3, line4, line5, line6], loc='best')
plt.xlabel( ' $\delta$/$\sigma$ ',fontsize=15)
plt.ylabel( 'Fisher Information/(N/$4\sigma^2$)',fontsize=15)
plt.show()
```

The figure generated by the above code denotes F($\delta,\gamma$) plotted for various $\gamma$ values which matches with the FI prediction made by the LS model. In particular, we also note that F($\delta,|\gamma|=1$)=0 for all values of $\delta$. Note that $\gamma=1$ curve is exactly on the top of $\gamma=-1$ curve in the above plot.

```python
import numpy as np
x=np.linspace(0.0001,5,100)
d=np.exp(-x**2/8)
Gamma=-1
y=(x**2)*(d**2)*(1-Gamma*d)/(4*(1-d**2)) #equation for the TN model 
line1,=plt.plot(x,y,label='$\gamma=-1$')
Gamma=-0.75
y=(x**2)*(d**2)*(1-Gamma*d)/(4*(1-d**2))
line2,=plt.plot(x,y,label='$\gamma=-0.75$')
Gamma=-0.5
y=(x**2)*(d**2)*(1-Gamma*d)/(4*(1-d**2))
line3,=plt.plot(x,y,label='$\gamma=-0.5$')
Gamma=0
y=(x**2)*(d**2)*(1-Gamma*d)/(4*(1-d**2))
line4,=plt.plot(x,y,label='$\gamma=0$')
Gamma=0.5
y=(x**2)*(d**2)*(1-Gamma*d)/(4*(1-d**2))
line5,=plt.plot(x,y,label='$\gamma=0.5$')
Gamma=1
y=(x**2)*(d**2)*(1-Gamma*d)/(4*(1-d**2))
line6,=plt.plot(x,y,label='$\gamma=1$')
plt.legend(handles=[line1, line2, line3, line4, line5, line6], loc='best')
plt.xlabel( ' $\delta$/$\sigma$ ',fontsize=15)
plt.ylabel( 'Fisher Information/(N/$4\sigma^2$)',fontsize=15)
plt.show()
```

The above plot for TN prediction does not follow from the calculations of $P_1$ and $P_2$ as they are not normalized. The FI plotted in the TN model is derived using $P_i (1+\gamma d)$ instead of $P_i$, although $\sum_{i=1,2} P_i (1+\gamma d)\neq 1$. Thus on plugging the revised values for $P_i$, we get the  FI for the parity sorter regarding the TN model as
\begin{equation*}
    F(\delta, \gamma)=\frac{\delta^2 d^2(1-\gamma d)}{16 \sigma^4 (1-d^2)}
\end{equation*}
Note that $\gamma=0$ curves is same for both the plots.

### 2.4 Results

FI measurement is two fold process
Firstly, Our goal is to calculate modal weights $P_{1,2}$ as a function of $\delta$ and $\gamma $, then we will use modal weights in above equation to calculate the values of FI.
On the second part we compare the variance in $\delta$ estimate with the FI prediction.
For the first part, 
for a fixed value of $\delta$ and corresponding antisymmetric mode, we record photon numbers $N_2$ at the output of parity sorter. Then we prepare symmetric mode for given $\delta$ and measure number of photons $N_1$ with different values of input powers.
The relation between $N_1$ and $N_2$ is:
$$ N_1 = \frac{(1+\gamma)(1+d)}{(1-\gamma)(1-d)}N_2$$
Now, we post select the value of $N_1$ corresponding to our desired $\gamma $. We calculate normalised modal weights as:
$$ P_{1,2} =\frac{ N_{1,2}}{N_1+N_2}$$


 eq(1)
$$ P_{1}(\delta , \gamma ) = \frac{(1+\gamma)(1+d)}{2(1+\gamma d)}  $$  
eq(2)

$$ P_{2}(\delta , \gamma ) = \frac{(1-\gamma)(1-d)}{2(1+\gamma d)} $$              
where d = $ e ^{\left(-\frac{\delta ^{2}}{8 \sigma ^{2}}\right)} $                            

For k = $ \delta / \sigma$

eq(3)
$$\frac{\partial P_{1}}{\partial \delta} = -\frac {kd(1-\gamma ^{2})}{8\sigma(1+\gamma d) ^{2}}  $$             
and
eq(4)
$$\frac{\partial P_{2}}{\partial \delta} =  \frac {kd(1-\gamma ^{2})}{8\sigma(1+\gamma d) ^{2}}  $$  



```python
import numpy as np
k = np.linspace(0.1,1)
sigma = 0.000327               ## in meters
delta = k*sigma          ## k = delta /sigma
p = np.square(k)/8       ## ((delta/sigma)^2)/8
d = np.exp(-p)


def plotf(gamma):
    p1 =((1+gamma)*(1+d))/(2*(1+gamma*d))               ## model weights for even mode
    p2 =((1-gamma)*(1-d))/(2*(1+gamma*d))               ## model weights for odd mode
    if gamma== -1:
        p4 = p2*0.92                                        ## model weights for 8% cross-talk for odd mode
        p3 = p1*0.92+0.08                                       ## model weights for 8% cross-talk for even mode
    else:                                              
        p4 = p2*0.92+0.08
        p3 = p1*0.92
    
    fig, ax = plt.subplots()
    ax.plot(k,p2,label = 'odd modes ideal')                   ## plot model weights for odd mode
    ax.plot(k,p1,label = 'even modes ideal')                  ## plot model weights for even mode
    line1, = ax.plot(k,p4, '--', linewidth=2,                 ## plot model weights for odd mode with 8% cross-talk
                 label='odd modes with cross talk')           
    line2, = ax.plot(k, p3, '--', linewidth=2,                ## plot model weights for even mode with 8% cross-talk
  
                 label='even modes with cross talk')

    ax.legend(loc='center left')
    plt.xlabel('$\delta/\sigma$')
    plt.ylabel('Modal weights')                            ## plot title for corresponding gamma values
    if gamma== -1:
        plt.title('$\gamma = -1$')
        plt.title('(a)', loc='left')
    elif gamma == -0.75:
        plt.title('$\gamma = -0.75$')
        plt.title('(b)', loc='left')
    elif gamma == 1:
        plt.title('$\gamma = 1$')
        plt.title('(c)', loc='left')
        
    plt.show()                                             ## show the plot
    
plotf(-1)                                                  ## call plotf function to plot for given gamma value
plotf(-0.75)
plotf(1)
        
        
```

FIG. 3: Modal weights for the parity sorter for $\gamma = -1,1,-0.75.$
Solid lines indicate theoretically expected modal weights with zero cross-talk while dashed lines indicate the expected modal weights for 8% cross-talk

FIG.3. shows modal weights with different $\gamma$ values. For $\gamma = +1,-1$ all the optical power goes either in even or odd port. When $\gamma = -1$, P1 = 0 and P2 = 1 and when $\gamma = 1$, P1 = 1 and P2 = 0. For $\gamma = +1,-1$, the values of modal weights are constant with respect to $\delta$ and hence we don't get any information from parity sorting.

From FIG. 3(b), we infer that modal weights changes quickly with respect to value $\delta$ which is responsible for higher FI value for $\gamma = -0.75$. In calculation of modal weights we have subtracted 8% of cross talk from it and plotted in FIG. 3. This cross talk can corresponds to any of the following reasons:
1. Intensity mismatch of the two beams
2. Finite coherence time
3. Fluctuations of path length
4. Polarization in the two arms

Better alignment can reduce cross-talk.




FI for parity sorter:
$$F(\delta , \gamma ) = \sum \limits _{i=1,2}  \frac{1}{P_{i}(\delta , \gamma )} \left(\frac{\partial P_{i}(\delta , \gamma )}{\partial \delta}\right)^ {2}$$ where $P_{1}(\delta , \gamma ),P_{2}(\delta , \gamma )$ are even and odd modal weights respectively.

```python
import numpy as np
def fi(gamma):
    
    if gamma ==1 or gamma == -1:
        fi =( k*k*d*d*(1-gamma*gamma))/(4*(1-d*d)*(1+gamma*d)*(1+gamma*d))
        return fi
    else:
        p1 =((1+gamma)*(1+d))/(2*(1+gamma*d))               ## model weights for even mode
        p2 =((1-gamma)*(1-d))/(2*(1+gamma*d))               ## model weights for odd mode
                                                          ## from eq 3 and 4
    derOfP2 = (k*d*(1-gamma*gamma))/(4*((1+gamma*d)**2))              ## derivative of odd modal weight times 2* sigma
    derOfP1 = -1* derOfP2                                             ## derivative of even modal weight times 2*sigma
    fi = (1/p1+ 1/p2)*((derOfP2)**2)                 ## calculation of FI/ 4*(sigma)^2
    return fi
                                                         

plt.plot(k,fi(-1),dashes=[5, 5, 5, 5],                    ## plot FI versus delta/sigma for different gamma values
                 label='$\gamma =-1$')  
plt.plot(k,fi(-0.75) ,label='$\gamma =-0.75$')

plt.plot(k,fi(-0.5),label='$\gamma =-0.5$')

plt.plot(k,fi(-0.25),label='$\gamma =-0.25$')
plt.plot(k,fi(0),dashes=[5, 5, 5, 5],
                 label='$\gamma = 0$')
plt.plot(k,fi(0.25),label='$\gamma = 0.25$')

plt.plot(k,fi(0.5),label='$\gamma = 0.5$')
plt.plot(k,fi(1),dashes=[10,10,10,10],
                 label='$\gamma = 1$')

plt.legend(loc='upper right')         
plt.xlabel('$\delta/\sigma$')                                   ## label the axis
plt.ylabel('Fisher Information / ( $N_{t}/4\sigma ^{2})$') 
plt.show()                                                      ## show the plot

```

## 3.Relation to previous works



We have shown above that with proper normalisation, the TN model and LS model agree with each other. We have verified experimentally that Rayleigh's curse reappears for $|\gamma| = 1$, as predicted in the LS model. We have also shown that per image plane photon, $\gamma < 0$ affords higher sensitivity than $\gamma = 0$, which has been predicted by both the TN and LS models. If we substitute $\gamma$ with $cos(\psi)$,  where $\psi$ is the relative shift between the two shifted Gaussian PSFs, our image plane FI results also agree with the predictions made by Hradil, Rehacek, Sanchez-Soto and Englert in their paper "Quantum fisher information with coherence".

Also, for $\gamma \rightarrow -1$ and $\delta \rightarrow 0$, the image plane power vanishes,and one needs to pump
harder at the source end to maintain a constant photon number at the image plane. Hradil et al., in their paper, correctly argues that this ‘cost’ of generating coherence should be taken into account when discussing the FI of partially coherent sources. We call this cost-accounted FI as the object plane FI, and assumes a constant photon number at the object plane. The object plane FI for $\gamma < 0$ can still exceed the $\gamma = 0$ case, however it cannot be higher than twice the $\gamma = 0$ case. Also, the range of $\delta$ for which the $\gamma < 0$ FI is higher than the $\gamma = 0$ FI grows smaller as $\gamma \rightarrow −1$. These results do not contradict the analysis of Hradil et al. since our source preparation model is different. In the notation of Hradil's paper, we assume the source $|\phi> = |\Phi_1> |\uparrow_z>$ can be prepared directly without deriving it as a subensemble of a classically entangled state, as done by Hradil et al. This mimics more closely the scalar theory problem of two point sources emitting in the same polarization,which is an assumption made in all of the preceding papers on superresolution and SPADE for incoherent  or partially coherent sources. Also, the object plane FI does not suffer from the ‘divergence’ predicted in the image plane FI for the case of vanishing power for $\gamma \rightarrow -1$ and $\delta \rightarrow 0$.

In a realistic scenario, $\gamma$ itself can be unknown and needs to be estimated jointly with $\delta$. Other potentially
unknown parameters include the phase of $\gamma$, the centroid, and the intensity ratio of the two sources, and we expect Rayleigh’s curse to persist in general. Finding the quantum FI and the optimal measurement scheme for this general scenario is infact an active area of research. In the special case of when only $\delta$ is unknown, our work shows how the lowerbound on the Var[$\hat{\delta}$] changes with $\gamma$ for a given photon budget in the image or object plane.




## 4.Conclusion



We performed parity sorting on two Gaussian PSFs with varying degrees of spatial coherence. Our results show that partial anticorrelation of the two sources increases the FI of $\delta$ estimation. For perfect coherence ($|\gamma| = 1$), the FI is zero and SPADE does not offer any advantange for sub-diffraction limit localization. The principal difference between the predictions of the TN model and LS model is due to the normalization of modal weights. We have shown that after proper normalization, the TN model gives the same FI predictions as the LS model. Therefore, Rayleigh’s curse can be avoided for partially coherent sources. We have assumed a real, known value of $\gamma$. As an extension of this work,the more realistic case of multiparameter estimation of a complex $\gamma$, and the centroid and intensity ratio of the two sources can be considered.
