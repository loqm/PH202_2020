# **Exploiting the speckle-correlation scattering matrix for a compact reference-free holographic image sensor** 

**THE ABSTRACT**



*   The word 'holography' means a drawing that contains all of the information for light - both amplitude and wavefront.
*   However, because of the insufficient bandwidth of current electronics, the direct measurement of the wavefront of light has not yet been achieved. Though reference-field-assisted **interferometric** methods have been utilized in numerous applications but, intoducing a reference field raises several fundamental and practical issues.
*   Here we demonstrate a reference-free holographic image sensor.

But, what exactly does it mean when we say **a reference field assisted holography** ?


*   A LASER beam is divided into **2** parts.
*   **Reference beam** reaches the photographic plate directly from the reflection mirror.
*   Waves in the **object beam** is reflected from different parts of the object.
*   Interference pattern formed by the superposition of coherent light beam is recorded on the photographic plate.
*   Intensity and phase both are recorded.

As we need to record the phase information as well, we need light waves to be in phase and hence a coherent beam is required. The best way is to use a LASER.

*   	Unfortunately, despite the significant efforts steadily continued for decades, the realizations and applications of holography are limited.
*  	One prime reason is the practical issues of introducing a reference field. It makes the system complex, incompatible and vulnerable to ambient noise.
*   However, although several reference-free holographic methods have been proposed, most of them have to sacrifice the generality of a holographic system by introducing specific assumptions about a sample or an incident beam.
*  Some examples of these are Shack-Hartmann-type sensors, transport- of-intensity equation, ptychographic scanning methods and iterative algorithms.

**THE MAIN IDEA**

The proposed idea is **a speckle- correlation scattering matrix (SSM)** approach; light-field information passing through a thin disordered layer is recorded and retrieved from a single-shot recording of speckle intensity patterns. Self-interference via diffusive scattering enables access to impinging light-field information, when light transport in the diffusive layer is precisely calibrated.

*  	Why SSM? - phase information is exploited in terms of **speckle correlation** and thus removes the need for reference beam.
*   The basic principle of our method is based on *the linear relationship between incident light and scattered light* from the diffusive layer, and *the randomly distributed (or complex Gaussian distributed) property of the diffused light*.
*   This whole linear relationship between incident field and the scattered field is described using **scattering matrix**, of which the transmission matrix (TM) is a subset.

**MATHEMATICAL ASPECTS**

*  For N orthonormal fields k1, k2, … . , kN as the preset input basis, the p-th column of
TM, 𝑡𝑝 , is the corresponding diffused field for the p-th input field 𝑘𝑝 .

*   Then, for a given arbitrary incident field,$ 𝑥 = \Sigma_p^n	𝛼_𝑝𝑘_𝑝$, the corresponding
diffused field can be expressed as $𝑦 = \Sigma_p^𝑛	𝛼_𝑝𝑡_𝑝$ , where $𝛼_1, 𝛼_2,… .. , 𝛼_𝑁$ are
complex-valued coefficients that contain the incident field information.
*   The camera captures the diffused intensity, y*y.

---
*$y*y$ loses the wavefront information during the capturing process and gives only the intensity speckle relation but we need intensity as well as wavefront information too! Hence, this info of wavefront is retrieved from the TM (it gives field-field correlation)*

---






*   Here, exploiting the random property of diffused fields, we find that calibrated TM, $t_1, t_2, …. .., t_N$, and the intensity snapshot, $y*y$, are sufficient for the incident field retrieval
*  Hence the proposed speckle-correlation scattering matrix, $Z$ is $–$
  


> $Z_pq=1\div(\Sigma_𝑝\Sigma_q)[ < 𝑡_𝑝∗𝑡_𝑞𝑦∗𝑦 >_𝑟	− < 𝑡_𝑝∗𝑡_𝑞 >_𝑟 < 𝑦∗𝑦 >_𝑟] \longrightarrow eqn.1$ 

> Here, $<.>_r$, is space average
>  and, $\sigma_p = <|𝑡𝑝|^2>_r$ is a normalisation constant.


*  The random property of diffused light field enables an intriguing mathematical relation between fourth order and second-order moments $–$


>  $< 𝐸_1𝐸_2𝐸_3𝐸_4 > = < 𝐸_1𝐸_2 >< 𝐸_3𝐸_4 > + < 𝐸_1𝐸_3 >< 𝐸_2𝐸_4 > + < 𝐸_1𝐸_4 >< 𝐸_2𝐸_3 > \longrightarrow eqn.2$ 
>  Here, $E_1,E_2,E_3$ and $E_4$ are the 
diffused light fields
>  Using equation $1$ and $2$, we get $–$

>  $Z_pq=\alpha_p\alpha_q^* + 1\div(\Sigma_p\Sigma_q)[< 𝑡^∗𝑦^∗ >_r< 𝑡_𝑞𝑦 >_r	] \longrightarrow eqn.3$

*   In the above equation, $𝛼_𝑝 =	1/\Sigma_𝑝<𝑡_p^∗𝑦>$	and from the general orthogonality between uncorrelated diffused fields,

	$1/(\Sigma_p\Sigma_q)< 𝑡_p^∗𝑡_q >= \delta_pq$

*  Although the assumption of orthogonality has generally been accepted for
transport through a diffusive layer, correlation exists for highly scattering media, according to **the random matrix theory**.
*  Nonetheless, we note that this correlation can be ignored for a diffusive layer, which is used in this study. If it exists, these correlated elements can be numerically filtered out once a TM is calibrated and its eigenbasis is analysed.
*  Because the second term in equation “3” comprises the spatial average of other random fields $𝑡_𝑝𝑦^∗$ and $𝑡_𝑞𝑦$, it vanishes as the ratio $(\gamma)$** between the number of optical **sampling modes (M)** and the number of **preset input modes (N)** increases. 
*  •	Therefore, for a sufficiently large “$\gamma$” , the second term of equation “3” vanishes, and the rank of $Z$ matrix becomes *unity*; $Z_pq=\alpha_p\alpha_q$, whose sole eigenvector is the incident field.

## *Cause of noise in the proposed method*



*   Using coherent illumination, a holographic register is made by the light interference of the reference beam and object beam coming from the transmission or reflection of an object. The holograms are recorded digitally to be numerically reconstructed, retrieving the amplitude and phase information. When coherent light strikes an object with rough surfaces ,scaled to the visible wavelength range, a coherent noise or speckle appears. The hologram stores this noise along with the object information, and its presence deteriorates the quality of reconstructed image.
*   **Speckle noise** is a common phenomenon in all coherent imaging systems like laser, acoustic and SAR imagery. Speckle noise is an undesirable effect. The source of this type of noise is random interference between the coherent returns issued from the so many scatterers present on the surface on the scale
of wavelength of the incident wave.

## **Initialization**

```matlab
gamma = 5*5; % M to N ratio.
SNR = 100; % linear signal-to-noise ratio (SNR) to mimic practical noisy situations.
x = rand(1024,1)
```

```matlab
save randominputfield.mat
```

## **Incident field**

```matlab
load('randominputfield.mat');
N = numel(x);
M = gamma*N;
x = x ./ norm(x); % normalize X for simplicity
X2d = reshape(x,[sqrt(N),sqrt(N)]) % 2D reshape.
```

## **Transmission matrix (TM)**

```matlab
disp('TM generate start!'); tic;
```

```matlab
TM = raylrnd(sqrt(1/2/M),M,N).*exp(1i * random('unif',-pi,+pi-eps('double'),M,N)); % transmission matrix generation.
TM_prac = awgn(TM,SNR,'measured','linear') % add noise upon preset SNR.
```

image.png

```matlab
disp(['TM generate end! ... Elapsed time: ', num2str(toc),'s']);
```

## **Intensity speckle generation**

```matlab
Y = TM * x; % diffused field y generation.
Y_prac = awgn(Y,SNR,'measured','linear'); % add noise upon preset SNR.
Y_prac2d = reshape(Y_prac,[sqrt(M),sqrt(M)]); % 2D reshape. #for visualization.
Iyy = abs(Y_prac).^2; % measured intensity speckle single shot.
Iyy2d_osampd = abs(fftshift(ifft2(ifftshift(...
    padarray(fftshift(fft2(ifftshift(Y_prac2d))),[2*sqrt(M),2*sqrt(M)]))))).^2 % oversample Iyy.
```

image.png

## **Field retrieval using SSM method**

```matlab
disp('initial guess start!'); tic;
```

```matlab
sigma = mean(abs(TM_prac).^2,1); % normalization factor, sigma (see Eq. (2)).
Zmatrix = (TM_prac'*(repmat(Iyy-mean(Iyy),[1,N]).*TM_prac))./(sigma'*sigma); % Z matrix calculation.
[Xretv0,~] = eigs(Zmatrix,1); % Field retrieval by selecting the eigenvector of the largest eigenvalue.
disp(['initial guess end! ... Elapsed time: ', num2str(toc),'s']);
```

image.png

```matlab
Xcorr0 = x'*Xretv0/norm(x)/norm(Xretv0); % calculate correlation between X and Xretv0.
Xretv0 = Xretv0*exp(-1i*angle(Xcorr0)); % global phase correction.
Xretv02d = reshape(Xretv0,[sqrt(N),sqrt(N)]) % 2D reshape.
```

image.png

# **Modified Gerchberg-Saxton algorithm**

```matlab
iterMax = 1000; % max iteration number to prevent infinite loop.
iterTor = 10^-5; % convergence criteria.
disp('inverse TM calculation start!'); tic;
```

```matlab
[U,S,V]=svd(TM_prac,'econ'); % economic svd for inverse TM calculation.
sv = diag(S);
TMinv_prac = V*(repmat(1./sv,[1,M]).*U'); % inverse TM calculation.
clearvars U S V
disp(['inverse TM calculation end! ... Elapsed time: ', num2str(toc),'s']);
```

image.png

```matlab
disp('Modified Gerchberg-Saxton iteration start!'); tic;
```

image.png

```matlab
Xcorr_iter = zeros(iterMax,1); % archives correlation changes during iteration.
Xiter = Xretv0; % set retrieved field for the initial guess.
Xcorr_iter(1) = Xcorr0;
for itr = 1:1:iterMax
    Yiter = TM_prac * Xiter; % transform incident field into corresponding diffused field space.
    Yiter = sqrt(Iyy).*exp(1i*angle(Yiter)); % replace amplitude part with measured intensity.
    Xiter = TMinv_prac*Yiter; % return to the incident field space.
    Xcorr_iter(itr+1) = x'*Xiter/norm(Xiter); % correlation calculation.
    if abs(Xcorr_iter(itr+1)) - abs(Xcorr_iter(itr)) < iterTor % iteration ends when the criterion is satisfied.
        break;
    end
end
disp(['Modified Gerchberg-Saxton iteration end! ... Elapsed time: ', num2str(toc),'s']);
```

image.png

```matlab
Xiter = Xiter*exp(-1i*angle(Xcorr_iter(itr))); % global phase correction. 
Xiter2d = reshape(Xiter,[sqrt(N),sqrt(N)])./ norm(Xiter); % 2D reshape.
```
