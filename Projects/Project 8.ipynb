# Introduction and limitations of a normal lens

A normal lens function is to apply phase correction to each of the components of a wave by refraction (for all wavelenghts) and focus them on a screen to produce a image of the object. Lens have been used throughout the world for focusing all light components since 424 BC, however in 17th and 18th it has been shown and mathematically proven too that a single element lens cannot focus all the components of the light effectively. Since then, opticians have tried to change their curvatures, wrongly assumed errors arosed from defects in the spherical figures of the surface, which led to creation of the compund achromatic lens in 1733 by Chester Moore Hall. Let's look at this limitation mathematically,

lets take a object in the X-Y direction , $\hat{O}(\textbf{r},t) = \sum\limits_{k_{x} k_{y}}  A_{k_{x} k_{y}} e^{i(x k_{x}+ y k_{y}+ z k_{z}- \omega t)}$

Now we know that by maxwell equations, we have relations between wave vectors and frequency of the wave by choosing axis of the image to be Z-axis,
                            $k_{z} = \sqrt{\omega^2 c^{-2} - k_{x}^2 - k_{y}^2}$

But for large values of wave vector $k_{x}^2 + k_{y}^2 > \omega^2 c^{-2}$, we will have,
                            $k_{z} = i\sqrt{k_{x}^2 + k_{y}^2 - \omega^2 c^{-2}}$
 
 Instead of being normal waves these waves will be,
$\hat{O}(k_{x}^2 + k_{y}^2 > \omega^2 c^{-2}) = \sum\limits_{k_{x}}\sum\limits_{k_{y}} A_{k_{x} k_{y}} e^{-k_{z}z} e^{i(x k_{x}+ y k_{y}- \omega t)}$  

$\textbf{A normal propogating wave in y-z direction}$

                               $\textbf{A normal evanescent wave in y-z direction}$


It is quiet evident that a evanescent wave will decay in the direction of axis of the lens and propagating waves would not.
These evascent waves are removed from the image as their ampltiude decays away by the factor of  $e^{-k_{z}z}$ in the Z-direction and it's not possible to focus the propagating waves in an area of less than wavelenght squared, which results is image consisting of only propogating waves, hence image cannot be focused in a area less than wavelenght squared and it looses it's details about the smallest and the minute details about the object.

The fact of having resolution of the order of wavelenght for the propagating waves, comes from the fact that, it is not possible for the device or your eyes to differtiate between two points of maximum brightness of the two images closer than almost the wavelength distance, since they superimpose and the image seems blurry from the laws of diffraction. # Meta Materials



A metamaterial is a macroscopic composite of periodic or non-periodic
structure, whose function is due to both the cellular architecture and the chemical composition. If the metamaterial is regarded as an effective medium, there is
an additional requirement that the cellular size is smaller than or equal to the subwavelength.

One such example[1] is that of a plasmon: a collective oscillation of electron density. In equilibrium the charge on the electron
gas is compensated by the background nuclear charge.
Displace the gas and a surplus of uncompensated charge is
generated at the ends of the specimen, with opposite signs
at opposite ends supplying a restoring force resulting in
simple harmonic motion,

${\omega_p}^2 =\frac{n{e}^2}{\epsilon_0 m_{eff}} \tag{1}$

The plasma frequency, $\omega_p$, is typically in the ultraviolet
region of the spectrum: around 15 eV in aluminum.
The plasmons have a profound impact on properties of
metals, not least on their interaction with electromagnetic
radiation where the plasmon produces a dielectric function
of the form

$\epsilon (\omega) =1 - \frac{{\omega_p}^2}{\omega (\omega + \iota\gamma)} \tag{2}$

which is approximately independent of wave vector,
and the parameter $\gamma$ is a damping term representing
dissipation of the plasmons energy into the system.

The significant point about Eq.(2) is that it is essentially negative below the plasma frequency, at least down
to frequencies comparable to $\gamma$.

There have been similar research for  tuning the $\mu$ of a material.In [2],the authors  have also shown that a structure containing loops of conducting wire has properties mimicking a magnetic plasma, 

$\mu \approx 1 - \frac{\omega_{mp}^2}{\omega^2} \tag{3}$

## Lets explore more
Now lets try to see how could macroscpic parameters such as cell size or structure, with some chemical composition help us to get a medium with negative refractive index[3].

We now derive the plasma frequency for collective oscillations of electrons in the wires. Consider a displacement of electrons along one of the cubic axes: The active wires will be those directed along that axis. If the density of electrons in these wires is n, the density of these active electrons in the structure as a whole is given by the fraction of space occupied by the wire,

$n_{eff} = n {\frac{\pi r^2}{a^2}}\tag{4}$

we also need to consider one thing : Any
restoring force acting on the electrons will not only have
to work against the rest mass of the electrons, but also
against self-inductance of the wire structure.. It
can be represented as a contribution to the electron mass.
Suppose a current
$I$ flows in the wire creating a magnetic field circling the
wire,

$H(R) = {\frac{\pi r^2 n \nu e}{2 \pi R}}\tag{5}$

where R is the distance from the wire center. We have
also reexpressed the current in terms of electron velocity,
$\nu$, and charge density, $ne$.We write the magnetic field in terms of a vector potential,

$\mathbf{H} (R) = {\frac{\nabla \times \mathbf{A} (R)}{\mu_0}}\tag{6}$

<img src="img2.PNG" width=400 height=400>

For an isolated wire the vector potential is ill-defined
until boundary conditions are specified. However, for a
three-dimensional array of wires, the mutual inductance
actually simplifies the problem and introduces the lattice
spacing as a natural cutoff,

$A(R) = \frac{\mu_0 \pi r^2 n \nu e}{2\pi} ln(\frac{a}{R})\tag{7}$

where a is the lattice constant.

We note that, from classical mechanics, electrons in
a magnetic field have an additional contribution to their
momentum of eA, and therefore the momentum per unit length of the wire is

$\pi r^2 e n A(r) = {\frac{\mu_0 \pi^2 r^4 e^2 n^2 \nu}{2\pi} } ln(\frac{a}{r}) = \small m_{eff} \pi r^2 n \nu \tag{8}$

where meff is the new effective mass of the electrons
given by

$m_{eff} = \frac{\mu_0 \pi  r^2 e^2  n}{2 \pi} ln(\frac{a}{r})\tag{9}$

Having both the effective density, neff, and the effective
mass, $m_{eff}$, on hand we can substitute into (1),

${\omega_p}^2 =\frac{n{e}^2}{\epsilon_0 m_{eff}} = \frac{2 \pi}{a^2 \epsilon_0  \mu_0} = \frac{2 \pi c_0}{a^2 ln(a/r)}\tag{10}$

where $c_0$ is speed of light in vaccum.

thus we expressed $\omega_p$ only in terms of lattice  constants.

How stable is the plasmon?

The plasmon may decay through electron-hole pair creation,
or through generation of phonons in the wires, depending
on the temperature. Either way, this mechanism acts
entirely through the electrical resistance of the wires. A
more careful calculation including resistance gives the
following expression for an effective dielectric function
of the structure:

$\epsilon_{eff} = 1 - \frac{{\omega_p}^2}{\omega (\omega \space +  \space i \epsilon_0 a^2 {\omega_p}^2/\pi r^2 \sigma )}\tag{11}$

where $\sigma$ is the conductivity of the metal.

A similar formation exists for tuning  $\mu$ of a material[4] but we shall  not be going there this time.

As we have shown it earlier. The propagating waves are limited
$\omega^2c^{-2} > k_y^2+k_x^2$

To check the validity we take  the following assumtion as $ \mu = -1 $ and $ \epsilon = -1 $. Thus we get refractive index as following:
$n = \sqrt{\mu*\epsilon}$

And to support these argument we define the impedance of the medium as
$ Z = \sqrt{\frac{\epsilon_{0}*\epsilon}{{\mu_{0}*\mu}}} $.
Thus impedance retains its positive sign so that, when both  $ \mu = -1 $ and $ \epsilon = -1 $ the medium is a perfect match to free space and the interfaces show no reflection.
$  $

Confirming that all of the energy is perfectly transmitted into the medium but in a strange manner: transport of energy in the positive $ k_{z}' $ direction requires that it takes the following value
 $ k_z' = -\sqrt{\omega^2c^{-2}-k_y^2-k_x^2}  $
 Finally, the transmission coefficient of the medium is
 $ T_S = tt'*exp(ik_z'd) = tt'*exp(-i\sqrt{\omega^2c^{-2}-k_y^2-k_x^2}d) $
 where d is the slab thickness and the negative phase results from the choice of wave vector forced upon us by causality.
 
All this was pointed out by Veselago [5] some time ago


Coming to the proof.
Let us assume the following electric field given in various directions.

$ E_{0S+} = [0, 1, 0]exp(ik_zz + ik_xx- i\omega t) $ ( S-polarized light in vacuum)

$ E_{0S-} = r[0, 1, 0]exp(ik_zz + ik_xx- i\omega t) $( reflected light )

$ E_{1S+} = t[0, 1, 0]exp(ik_z'z + ik_xx- i\omega t) $ (while some transmited)
  
  Here we can define $ k_z $  and $ k_z' $ respectively,
  
   $ k_z = +i\sqrt{\omega^2c^{-2}-k_y^2-k_x^2}  $,
  for $ \omega^2c^{-2}<k_y^2+k_x^2 $
  
 $ k_z' = +i\sqrt{\epsilon\mu\omega^2c^{-2}-k_y^2-k_x^2},  $ for $ \epsilon\mu\omega^2c^{-2}<k_y^2+k_x^2$
 

By matching wave fields at the interface, from meduim to lens and conversely. We get the coefficient as the following:

$ t = \frac{2 k_z'}{\mu k_z + k_z'}  $
  $ r = \frac{\mu k_z - k_z'}{\mu k_z + k_z'} $
  
 $ t' = \frac{2 k_z'}{\mu k_z + k_z'} $
  $ r' = \frac{k_z' - \mu k_z }{ k_z'+\mu k_z } $  
  
  
Let's calculate transmission coefffiecnt through both surfaces of the slab. We must sum the multiple scattering events  
$ T_S = tt'exp(ik_z'd) + tt'r'^(2)exp(3ik_zd) + tt'r'^(4)exp(5ik_zd).... $

Summing the above geometric system

$ T_S = tt'\frac{exp(ik_z'd)}{1-r'^2exp(2ik_z'd)} $ 


  
  
  
Taking the limiting case for T_S:

$\begin{aligned} \lim _{\mu \rightarrow-1,\epsilon \rightarrow-1} T_{S}=& \lim _{\mu \rightarrow-1,\epsilon \rightarrow-1} \frac{2 \mu k_{z}}{\mu  k_{z}+k_{z}^{\prime}} \frac{2 k_{z}^{\prime}}{k_{z}^{\prime}+\mu k_{z}} \\ & \times \frac{\exp \left(i k_{z}^{\prime} d\right)}{1-\left(\frac{k_{z}^{\prime}-\mu  k_{z}}{k_{z}^{\prime}+\mu  k_{z}}\right)^{2} \exp \left(2 i k_{z}^{\prime} d\right)} \\=& \exp \left(-i k_{z} d\right)=& \exp \left(-i k_{z}' d\right) \end{aligned}$  

Similarly for reflection coefficient

$\lim _{\mu \rightarrow-1 \atop \varepsilon \rightarrow-1} R_{S}=\lim _{\mu \rightarrow-1 \atop \varepsilon \rightarrow-1} r+\frac{t t^{\prime} r^{\prime} \exp \left(2 i k_{z}^{\prime} d\right)}{1-r^{\prime 2} \exp \left(2 i k_{z}^{\prime} d\right)}=0$

The result also holds for P-polarized evanescent waves:

$\begin{aligned} \lim _{\mu \rightarrow-1,\epsilon \rightarrow-1} T_{P}=& \lim _{\mu \rightarrow-1, \epsilon \rightarrow-1} \frac{2 \varepsilon k_{z}}{\varepsilon k_{z}+k_{z}^{\prime}} \frac{2 k_{z}^{\prime}}{k_{z}^{\prime}+\varepsilon k_{z}} \\ & \times \frac{\exp \left(i k_{z}^{\prime} d\right)}{1-\left(\frac{k_{z}^{\prime}-\varepsilon k_{z}}{k_{z}^{\prime}+\varepsilon k_{z}}\right)^{2} \exp \left(2 i k_{z}^{\prime} d\right)} \\=& \exp \left(-i k_{z}' d\right)=& \exp \left(-i k_{z} d\right) \end{aligned}$  
  
Conclusion:
Even though we have meticulously carried through a strictly causal calculation, our final result is that the medium does amplify evanescent waves. Hence we conclude that with this new lens both propagating and evanescent waves  contribute to the resolution of the image. Thus conclude that there is no physical obstacle to perfect reconstruction of the image beyond practical limitations of apertures and perfection of the lens surface.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
# Plot on that set of axes
x= np.linspace(-5*10**(-7),5*10**(-7),100)
lam= 3.57*(10**-7)
h1=8*10**(-8)
h2=4*10**(-8)
A = 10**5
d=1.75*10**(-7)  # half the distance between the objects
a=5*10**(-8)  # lenght of the objects
B1 = np.pi*a*(x-d)/(lam*h1)
B2 = np.pi*a*(x+d)/(lam*h1)
R1 = np.pi*a*(x-d)/(lam*h2)
R2 = np.pi*a*(x+d)/(lam*h2)
kx = 10*10**7
dec = np.exp(-(h1*kx)) #decaying factor
print(dec)
y1= 130*dec*dec*A*pow(np.sin(B1)/B1,2)
y2= 130*dec*dec*A*pow(np.sin(B2)/B2,2)
f1= A*pow(np.sin(R1)/R1,2)
f2= A*pow(np.sin(R2)/R2,2)
axes.plot(x, (y1+y2), 'b', label ='Image of the propogating waves formed by lens')
axes.plot(x, pow((dec/(0.04+dec)),2)*(f1+f2), 'r', label='Image of the double focused image of evanescent waves')
axes.set_xlabel('X-Axis (100nm)') # Notice the use of set_ to begin methods
axes.set_ylabel('Image Intensity at Z=80nm')
axes.set_title('Set Title')
plt.legend(bbox_to_anchor =(1.75, 1.25), ncol = 2)
fig,ax=plt.subplots(1,1)
y= np.linspace(-7*10**(-7),7*10**(-7),100)
x= np.linspace(-7*10**(-7),7*10**(-7),100)
X, Y = np.meshgrid(x, y)
C1 = np.pi*a*(Y)/(lam*h2)
C2 = np.pi*a*(Y)/(lam*h2)
A1 = np.pi*a*(-Y)/(lam*h1)
A2 = np.pi*a*(-Y)/(lam*h1)
D1 = np.pi*a*(X-d)/(lam*h1)
D2 = np.pi*a*(X+d)/(lam*h1)
G1 = np.pi*a*(X-d)/(lam*h2)
G2 = np.pi*a*(X+d)/(lam*h2)
z1= A*pow(np.sin(G1)/G1,2)*pow(np.sin(C1)/C1,2)
z2= A*pow(np.sin(G2)/G2,2)*pow(np.sin(C2)/C2,2)
v1= A*pow(np.sin(D1)/D1,2)*pow(np.sin(A1)/A1,2)
v2= A*pow(np.sin(D2)/D2,2)*pow(np.sin(A2)/A2,2)
Z1=v1+v2
cp = ax.contourf(X, Y,dec*dec*Z1,15)
fig.colorbar(cp)
fig,ax=plt.subplots(1,1)
ax.set_title('With the normal lens')
ax.set_xlabel('x (100nm)')
ax.set_ylabel('y (100nm)')
Z2=z1+z2
dp = ax.contourf(X, Y, pow((dec/(0.04+dec)),2)*Z2,15)
fig.colorbar(dp)
ax.set_title('With the silver slab')
ax.set_xlabel('x (100nm)')
ax.set_ylabel('y (100nm)')
Z= Z1 + pow((dec/(0.04+dec)),2)*Z2
plt.show()
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, dec*dec*Z1,50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, pow((dec/(0.04+dec)),2)*Z2, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```

Let's discuss about some weirdness in the negative refraction system,

$\textbf{1. Wave vector getting reversed inside the material??}$
This is a very intresting and a weird thing, because it makes us seem that two light waves, are going towards each other, and not that it is propagating!!
From our perspective this seems the case, in the perspective of the light its not travelling at all.

It's just like it is there in the same place, let's understand this
$\textbf{Light takes the shortest possible path between two points}$

Here we see that the shortest path is $D_{s}=n_{1}d_{1}+n_{2}d_{2}+n_{3}d_{3}$
and we know that $2d_{1}=2d_{3}=d_{2}$ in the case of slab $n_{2}=-1$, and the rest is air
so we get $D_{s}=0$, this result tells us that it's like someone created a sort of wormhole between two point and light is just there a two places not travelling at all!!

$\textbf{ What about energy??}$

Well evanscent waves doesn't actually carry the energy, but something weird happens here,
The direction of propagation of wave vector and the energy transportation are completely opposite to each other. Such materials are also called left hand material instead of usual right hand material.




It is basically impossible to recreate the exact image of the object becuase the medium isn't exaclt pure negative index type, there is going to be some dissapation in form or the other. With the help of the weird properties of the slab we have been able to overcome the limitations of the normal lens.
